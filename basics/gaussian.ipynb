# Gaussian
Some basics about the Gaussian function.
$$ \mu $$ is the mathematical expectation.
$$ \sigma $$ is the standard deviation.

$$ f(x,\sigma, \mu) = \frac {1} {\sigma\sqrt{2\pi}} e^{\frac {-(x-\mu)^2} {2}} $$


## Parameters
What happens if we make the parameters mu and sigma varying ?

```python
import numpy as np
%run lib/drawing.py

gauss = lambda x, sigma, mu: 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x-mu)**2 / 2)

X = np.arange(-4, 4, 0.1)
Y = gauss(X, 1, 0)

draw = Drawing(1, 1, xlim=(-4, 4), ylim=(0, 0.4))
draw.plot(X, Y, 'gaussian function', color='violet')
draw.show()

```

### Variation of the standard deviation

```python
import numpy as np
%run lib/drawing.py

gauss = lambda x, sigma, mu: 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x-mu)**2 / 2)

X = np.arange(-4, 4, 0.1)
Y = gauss(X, 1, 0)
Y_sigma2 = gauss(X, 2, 0)
Y_sigma3 = gauss(X, 3, 0)

draw = Drawing(1, 1, xlim=(-4, 4), ylim=(0, 0.4))
draw.plot(X, Y, 'gaussian function', color='violet')
draw.plot(X, Y_sigma2, 'gaussian function', color='red')
draw.plot(X, Y_sigma3, 'gaussian function', color='orange')
draw.show()
```

### Variation of the mathematical expectation 

```python
import numpy as np
%run lib/drawing.py

gauss = lambda x, sigma, mu: 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x-mu)**2 / 2)

X = np.arange(-4, 4, 0.1)
Y = gauss(X, 1, 0)
Y_mu1 = gauss(X, 1, 1)
Y_mu2 = gauss(X, 1, 2)

draw = Drawing(1, 1, xlim=(-4, 4), ylim=(0, 0.4))
draw.plot(X, Y, 'gaussian function', color='violet')
draw.plot(X, Y_mu1, 'gaussian function', color='red')
draw.plot(X, Y_mu2, 'gaussian function', color='orange')
draw.show()
```

### Inverse

```python
import numpy as np
%run lib/drawing.py

gauss = lambda x, sigma, mu: 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x-mu)**2 / 2)

X = np.arange(-4, 4, 0.1)
Y = gauss(X, 1, 0)
Y_inv = -gauss(X, 1, 0)

draw = Drawing(1, 1, xlim=(-4, 4), ylim=(-0.4, 0.4))
draw.plot(X, Y, 'gaussian function', color='violet')
draw.plot(X, Y_inv, 'gaussian function', color='orange')
draw.show()
```

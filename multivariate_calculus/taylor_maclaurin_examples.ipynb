## cosine
Let's have a look on the cosine function. 
Because we will look at the point x=0, we will use the MacLaurin series. 

> MacLaurin series
> $$ f(x) = \sum_{n=0}^{\infty} \frac {f^{(0)}(x)} {n!} x^{n} $$

$$ f(x) = cos(x) $$

### derivatives
In order to be able to apply TS series, we need to calculate the derivatives of f(x). We will study that function until the 16th approximation order. 

$$ f^{(1)}(x) = -sin(x)  $$
$$ f^{(2)}(x) = -cos(x)  $$
$$ f^{(3)}(x) = sin(x)  $$
$$ f^{(4)}(x) = cos(x)  $$
$$ f^{(5)}(x) = -sin(x)  $$
$$ f^{(6)}(x) = -cos(x)  $$
$$ f^{(7)}(x) = sin(x)  $$
$$ f^{(8)}(x) = cos(x)  $$
$$ f^{(9)}(x) = -sin(x)  $$
$$ f^{(10)}(x) = -cos(x)  $$
$$ f^{(11)}(x) = sin(x)  $$
$$ f^{(12)}(x) = cos(x)  $$
$$ f^{(13)}(x) = -sin(x)  $$
$$ f^{(14)}(x) = -cos(x)  $$
$$ f^{(15)}(x) = sin(x)  $$
$$ f^{(16)}(x) = cos(x)  $$

And the values at 0 for each is 
$$ f^{(1)}(0) = 0  $$
$$ f^{(2)}(0) = -1  $$
$$ f^{(3)}(0) = 0  $$
$$ f^{(4)}(0) = 1  $$
$$ f^{(5)}(0) = 0  $$
$$ f^{(6)}(0) = -1  $$
$$ f^{(7)}(0) = 0  $$
$$ f^{(8)}(0) = 1  $$
$$ f^{(9)}(0) = 0  $$
$$ f^{(10)}(0) = -1  $$
$$ f^{(11)}(0) = 0  $$
$$ f^{(12)}(0) = 1  $$
$$ f^{(13)}(0) = 0  $$
$$ f^{(14)}(0) = -1  $$
$$ f^{(15)}(0) = 0  $$
$$ f^{(16)}(0) = 1  $$


```python
X = np.arange(-8, 8, 0.1)
Y = np.cos(X)

draw = Drawing(1, 1)
draw.plot(X, Y, 'cosine', color='violet')
draw.show()
```

### approximations
$$ g_{0} = f(0) = 1 $$
$$ g_{2} = f(0) + \frac {f^{(2)}(0)} {2!} x^{2} = 1 - \frac {x^{2}} {2!}  $$
$$ g_{4} = f(0) + \frac {f^{(2)}(0)} {2!} x^{2} + \frac {f^{(4)}(0)} {4!} x^{4} = 1 - \frac {x^{2}} {2!} + \frac {x^{4}} {4!} $$
$$ ... $$
$$ g_{16} = f(0) + \frac {f^{(2)}(0)} {2!} x^{2} + ... + \frac {f^{(16)}(0)} {16!} x^{16} = 1 - \frac {x^{2}} {2!} + ... + \frac {x^{16}} {16!} $$

> Because the values at 0 for each odd derivative is 0, that derivatives are ignored.
> $$ g_{1} = f(0) + f^{(1)}(0)x $$
> $$ g_{1}(0) = 1 = g_{0}(0) $$

```python
import numpy as np
import scipy.misc

gn = lambda nx: np.array([np.power(-1,(n % 4) / 2) * np.power(nx[1],n) / scipy.special.factorial(n) for n in np.arange(0, nx[0]+1, 2)]).sum()

xs = np.arange(-8, 8, 0.1)
g0s = [gn((0, x)) for x in xs]
g2s = [gn((2, x)) for x in xs]
g4s = [gn((4, x)) for x in xs]
g6s = [gn((6, x)) for x in xs]
g8s = [gn((8, x)) for x in xs]
g10s = [gn((10, x)) for x in xs]
g12s = [gn((12, x)) for x in xs]
g14s = [gn((14, x)) for x in xs]
g16s = [gn((16, x)) for x in xs]

draw = Drawing(5, 2)
draw.plot(xs, Y, 'g0s', color='pink', index=1)
draw.plot(xs, g0s, 'g0s', color='violet', index=1)
draw.plot(xs, Y, 'g0s', color='pink', index=2)
draw.plot(xs, g2s, 'g2s', color='violet', index=2)
draw.plot(xs, Y, 'g0s', color='pink', index=3)
draw.plot(xs, g4s, 'g4s', color='violet', index=3)
draw.plot(xs, Y, 'g0s', color='pink', index=4)
draw.plot(xs, g6s, 'g6s', color='violet', index=4)
draw.plot(xs, Y, 'g0s', color='pink', index=5)
draw.plot(xs, g8s, 'g8s', color='violet', index=5)
draw.plot(xs, Y, 'g0s', color='pink', index=6)
draw.plot(xs, g10s, 'g10s', color='violet', index=6)
draw.plot(xs, Y, 'g0s', color='pink', index=7)
draw.plot(xs, g12s, 'g12s', color='violet', index=7)
draw.plot(xs, Y, 'g0s', color='pink', index=8)
draw.plot(xs, g14s, 'g14s', color='violet', index=8)
draw.plot(xs, Y, 'g0s', color='pink', index=9)
draw.plot(xs, g16s, 'g16s', color='violet', index=9)
draw.show()
```

### observation
We can see on the graphs above that each step improve the approximation of the cosine function.

# Newton-Raphson
NR method is useful to solve an equation and found a solution to the y=0 problem just by evaluate and differentiate the function.

$$ x_{n+1} = x_{n} - \frac {f(x_{n})} {f'(x_{n})} $$

## Example
Let's have a look on the following function:

$$ f(x) = \frac {x^6} {6} - 3x^4 - \frac {2x^3} {3} + \frac {27x^2} {2} + 18x - 30 $$ 

```python
%run lib/drawing.py
import numpy as np
f = lambda x: x ** 6 / 6 - 3 * x ** 4 - 2 * x ** 3 / 3 + 27 * x ** 2 / 2 + 18 * x - 30

X = np.arange(-8, 8, 0.1)
Y = f(X)

draw = Drawing(1, 1, xlim=(-8, 8), ylim=(-100, 100))
draw.plot(X, Y, 'f(x)', color='violet')
draw.show()
```

### derivative 
In order to use the NR method, we need to know the derivative form of the function:
$$ f'(x) = x^5 - 12x^3 - 2x^2 + 27x +18 $$

```python
d_f = lambda x: x**5 -12*x**3 -2*x**2 +27*x +18
```

### root calculation
in order to calculate the root near a point of the function, we use the code below. 

```python
import pandas as pd

def newton_raphson(x, f, d_f):
    d = {"x" : [x], "f(x)": [f(x)], "d_f(x)": [d_f(x)]}
    
    for i in range(0, 20):
      x = x - f(x) / d_f(x)
      d["x"].append(x)
      d["f(x)"].append(f(x))
      d["d_f(x)"].append(d_f(x))

    return pd.DataFrame(d, columns=['x', 'f(x)', 'd_f(x)'])
```

#### root near x = -4

```python
nr_table = newton_raphson(-4, f, d_f)
```

We observe that the function converge quickly to **-3.760214**.
We trace that point and it's gradient on a graph to figure out what's going on.

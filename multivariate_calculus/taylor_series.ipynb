# Taylor series

One of a family of approaches used for building approximation function. 


## Toolkit

```python
import matplotlib.pyplot as plt
import numpy as np

def init_drawing():
    """ Set axes and figure styles """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(1, 1, 1)
    fig.patch.set_facecolor('violet')
    fig.patch.set_alpha(0.0)
    ax.patch.set_facecolor('none')
    
    medium_grey = (0.4, 0.4, 0.4)

    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color(medium_grey)
    ax.spines['bottom'].set_color(medium_grey)
    ax.tick_params(axis='x', colors=medium_grey)
    ax.tick_params(axis='y', colors=medium_grey)
    fig.tight_layout()

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.set_facecolor((0.1, 0.1, 0.1))
```

Taylor series are also referred to as **power series**. And this is because they are composed of coefficients in front of increasing powers of x. So we can write a simple generalised expression for a power series as g of x, equals a, plus bx, plus cx squared, plus dx cubed et cetera. Potentially, going off for infinitely many times depending on what function we're considering.

$$ g_{0} = a $$ 


## Power series approximation
$$ g(x) = a + bx + cx^{2} + dx^{3} + ... $$ 

Starting from just a single term, we call these expressions the zeroth, first, second, and third order approximations etc. 

$$ g_{0}(x) = a $$ 

$$ g_{1}(x) = a + bx $$ 

$$ g_{2}(x) = a + bx + cx^{2} $$ 

$$ ... $$

## Example: a sine
Regarding a selected point of the curve, we will have a look on the approximation order representations to figure out how that lead us to a better approximation of the sine. 

$$ f(x) = sin(x) + 0.5 $$

### First approximation order
The first approximation order has a constant value on the y-axis passing through the point.

$$ g_{0} = a $$

```python
# the picked point
picked_point = (0.5, f(0.5))

# the first order approximation
line_g_0 = ([np.min(f_xs), np.max(f_xs)], [picked_point[1], picked_point[1]])

# draw the lines
init_drawing()

plt.scatter(picked_point[0], picked_point[1], color='yellow')
plt.plot(line_g_0[0], line_g_0[1], color='orange')
plt.plot(xs, ys, color='violet')
plt.show()
```

### Second order approximation
The second order approximation is a tangent passing through the picked point. The tangent is calculated using gradient. 
$$ g_{1}(x) = a + bx $$

```python
d_f = lambda x: np.cos(x)
line_g_1 = [[picked_point[0], picked_point[0] + 0.5], [picked_point[1], picked_point[1] + d_f(0.5)]]

# draw the lines
init_drawing()

plt.scatter(picked_point[0], picked_point[1], color='yellow')
plt.plot(line_g_1[0], line_g_1[1], color='blue')
plt.plot(line_g_0[0], line_g_0[1], color='orange')
plt.plot(xs, ys, color='violet')
plt.show()
```

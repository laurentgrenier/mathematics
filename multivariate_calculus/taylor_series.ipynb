# Taylor series

One of a family of approaches used for building approximation function. 


## Toolkit

```python
import matplotlib.pyplot as plt
import numpy as np

def init_drawing():
    """ Set axes and figure styles """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(1, 1, 1)
    fig.patch.set_facecolor('violet')
    fig.patch.set_alpha(0.0)
    ax.patch.set_facecolor('none')
    
    medium_grey = (0.4, 0.4, 0.4)

    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color(medium_grey)
    ax.spines['bottom'].set_color(medium_grey)
    ax.tick_params(axis='x', colors=medium_grey)
    ax.tick_params(axis='y', colors=medium_grey)
    fig.tight_layout()

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.set_facecolor((0.1, 0.1, 0.1))
```

Taylor series are also referred to as **power series**. And this is because they are composed of coefficients in front of increasing powers of x. So we can write a simple generalised expression for a power series as g of x, equals a, plus bx, plus cx squared, plus dx cubed et cetera. Potentially, going off for infinitely many times depending on what function we're considering.

$$ g_{0} = a $$ 


## Power series approximation
$$ g(x) = a + bx + cx^{2} + dx^{3} + ... $$ 

Starting from just a single term, we call these expressions the zeroth, first, second, and third order approximations etc. 

$$ g_{0}(x) = a $$ 

$$ g_{1}(x) = a + bx $$ 

$$ g_{2}(x) = a + bx + cx^{2} $$ 

$$ ... $$

## Example: Maclaurin approximation on an exponential
Regarding a selected point of the curve, we will have a look on the approximation order representations to figure out how that lead us to a better approximation of the exponential function. 

$$ f(x) = e^x $$

### Second order approximation
The second order approximation is a tangent passing through the picked point. The tangent is calculated using gradient. 
$$ g_{1}(x) = f(0) + f^{(1)}(0)x $$

```python
d_f_one = lambda x: np.exp(x)
g_one = lambda x: f(0) + d_f_one(0) * x

# line_g_1 = [[picked_point[0], picked_point[0] + 0.5], [picked_point[1], picked_point[1] + d_f(0.5)]]
g_one_ys = g_one(xs)

# draw the lines
init_drawing()

plt.scatter(picked_point[0], picked_point[1], color='yellow')
plt.plot(xs, g_one_ys, color='yellow', linestyle='dashed')
plt.plot(line_g_0[0], line_g_0[1], color='orange', linestyle='dashed')
plt.plot(xs, ys, color='violet')
plt.show()
```

### Third order approximation
$$ g_{2}(x) = f(0) + f^{(1)}(0)x + \frac{f^{(2)}(0)}{2}x^{2} $$

```python
d_f_two = lambda x: np.exp(x)

g_two = lambda x: f(0) + d_f_one(0) * x + (d_f_two(0)/2) * np.power(x, 2)
g_two_ys = g_two(xs)

# draw the lines
init_drawing()

plt.scatter(picked_point[0], picked_point[1], color='yellow')
plt.plot(xs, g_two_ys, color='pink', linestyle='dashed')
plt.plot(xs, g_one_ys, color='yellow', linestyle='dashed')
plt.plot(line_g_0[0], line_g_0[1], color='orange', linestyle='dashed')
plt.plot(xs, ys, color='violet')
plt.show()
```

### Fourth order approximation
$$ g_{3}(x) = f(0) + f^{(1)}(0)x + \frac{f^{(2)}(0)}{2}x^{2} + \frac{f^{(3)}(0)}{6}x^{3} $$

```python
d_f_third = lambda x: np.exp(x)

g_third = lambda x: f(0) + d_f_one(0) * x + (d_f_two(0)/2) * np.power(x, 2) + (d_f_third(0)/6) * np.power(x, 3)
g_third_ys = g_third(xs)

# draw the lines
init_drawing()

plt.scatter(picked_point[0], picked_point[1], color='yellow')
plt.plot(xs, g_third_ys, color='green', linestyle='dashed')
plt.plot(xs, g_two_ys, color='pink', linestyle='dashed')
plt.plot(xs, g_one_ys, color='yellow', linestyle='dashed')
plt.plot(line_g_0[0], line_g_0[1], color='orange', linestyle='dashed')
plt.plot(xs, ys, color='violet')
plt.show()
```
